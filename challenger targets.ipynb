```python
import plotly.express as px
import pandas as pd
from collections import Counter, namedtuple
pd.set_option('display.max_colwidth', -1)
```

```python
with open('challenger_targets') as f:
    data = f.read()
```

```python
Target = namedtuple('Target', ['challenger_name', 'challenger_loc', 'target_name', 'target_loc', 'score', 'distance'])
```

```python
targets = [Target(*x.split(",")) for x in [i.rstrip(".") for i in data.replace('''"''', "").split("\n")][:-1]]
```

```python
df = pd.DataFrame(targets, columns=Target._fields)
df['distance'] = df['distance'].astype(float)
pd.set_option('display.max_rows', df.shape[0]+1)
```
