```python
import json
from collections import defaultdict
import networkx as nx
import pandas as pd
```

```python
with open('files/boston.json') as json_file:
    data = json.load(json_file)
```

```python
data['links']
```

```python
df = pd.DataFrame(data['links'])
```

```python
df.head()
```

```python
graph = defaultdict(list)
```

```python
for x in data['links']:
    src = x['source']
    target = x['target']
    graph[src].append(target)
```

```python
df.to_csv(r'files/boston.csv', index=False)
```

```python
pd.DataFrame(graph.keys(), columns =['label']).to_csv(r'files/boston_nodes.csv', index=False)
```

```python
nodes = list(graph.keys())
```

```python
edges = []
for node in nodes:
    neighbors = graph[node]
    for n in neighbors:
        if n in nodes:
            edges.append({'source': nodes.index(node), 'target': nodes.index(n)})
```

```python
edges
```

```python
pd.DataFrame(edges).to_csv(r'files/boston_edges.csv', index=False)
```

```python
edges = []
for node in nodes:
    neighbors = graph[node]
    for n in neighbors:
        edges.append({'source': node, 'target': n, 'label': node})
```

```python
edges
```

```python
pd.DataFrame(edges).to_csv(r'files/boston_edges.csv', index=False)
```
