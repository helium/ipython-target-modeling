```python
with open('/tmp/foo', 'r') as f:
    data = f.read()
```

```python
d = eval(data.replace(' =>', ':').replace('#', '').rstrip('.\n'))
```

```python
for value in d.values():
    for v in value:
        print('{0:30} {1}'.format(v[0], v[1]))
```
