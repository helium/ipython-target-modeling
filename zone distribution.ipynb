```python
import plotly.express as px
import pandas as pd
from collections import Counter, namedtuple
import matplotlib
```

```python
with open('/tmp/hexes') as f:
    data = f.read()
```

```python
hexes = [eval(i.replace('{', '(').replace('}', ')').rstrip(',').replace('[', '').replace(']', '').strip())
         for i in data.split('\n')[:-1]]
```

```python
hexdict = dict(hexes)
```

```python
s = sum(hexdict.values())
p_hexdict = dict(map(lambda x: (x[0], x[1]/s), hexdict.items()))
```

```python
with open('/tmp/zones_1ZnhESdqLwJRvzEDwnVTqY6bMWdaunczduV8NBUUT9DDJAv3FD') as f:
    data = f.read()
```

```python
zones = [int(i) for i in data.split('\n')[:-1]]
```

```python
c = Counter(zones)
```

```python
df = pd.DataFrame.from_dict(c, orient='index').reset_index()
df = df.rename(columns={'index': 'zone', 0:'selection_count'})
```

```python
df['hex_count'] = df['zone'].map(hexdict)
```

```python
df['count_per_hex'] = df['hex_count']/df['selection_count']
```

```python
df['hex_prob'] = df['zone'].map(p_hexdict)
```

```python
df.sort_values('selection_count', ascending=False).head(50)
```

```python
px.histogram(df, x="hex_prob")
```

```python
px.histogram(df, x='selection_count')
```
